```python
#cutomer segmentation
from sklearn.preprocessing import StandardScaler

# Select relevant columns
df_selected = df[['MntTotal', 'Recency', 'NumWebVisitsMonth']]

# Normalize/Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_selected)
```

```python
import numpy as np

# Define the number of clusters (k)
k = 3

# Choose k random points as initial centroids
random_indices = np.random.choice(df_scaled.shape[0], size=k, replace=False)
centroids = df_scaled[random_indices]
```

```python
import numpy as np

# Define the number of clusters (k)
k = 3

# Choose k random points as initial centroids
random_indices = np.random.choice(df_scaled.shape[0], size=k, replace=False)
centroids = df_scaled[random_indices]
```

```python
from scipy.spatial.distance import cdist

# Calculate distances from each customer to each centroid
distances = cdist(df_scaled, centroids, metric='euclidean')
```

```python
# Assign each customer to the nearest centroid
cluster_assignments = np.argmin(distances, axis=1)
```

```python
# Update centroids based on the mean of assigned customers
new_centroids = np.array([df_scaled[cluster_assignments == i].mean(axis=0) for i in range(k)])
```

```python
# Iterate until convergence or a set number of iterations
max_iterations = 100
for i in range(max_iterations):
    distances = cdist(df_scaled, new_centroids, metric='euclidean')
    new_cluster_assignments = np.argmin(distances, axis=1)
    
    # Check if assignments have changed
    if np.array_equal(new_cluster_assignments, cluster_assignments):
        break
    
    cluster_assignments = new_cluster_assignments
    new_centroids = np.array([df_scaled[cluster_assignments == j].mean(axis=0) for j in range(k)])

# Final cluster assignments
df['Cluster'] = cluster_assignments
```

```python
# Visualize clusters (assuming 2D data)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_scaled[:, 0], y=df_scaled[:, 1], hue=df['Cluster'], palette='viridis')
plt.title('Customer Segmentation (K-means)')
plt.xlabel('MntTotal (Scaled)')
plt.ylabel('Recency (Scaled)')
plt.show()
```

```python
# Assuming 'df' is the dataframe and 'cluster_labels' is the series with cluster labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Income'], y=df['NumWebVisitsMonth'], hue=df['Cluster'], palette='viridis', s=100, alpha=0.7)
plt.title('Customer Segmentation: Income vs NumWebVisitsMonth')
plt.xlabel('Income')
plt.ylabel('Number of Web Visits per Month')
plt.legend(title='Cluster', loc='upper right')
plt.show()
```

```python
# Calculating the mean values of key features for each cluster
avg_values = df.groupby('Cluster')[['Income', 'MntTotal', 'NumWebVisitsMonth']].mean().reset_index()

# Plotting the bar chart
avg_values.set_index('Cluster').plot(kind='bar', figsize=(12, 6), colormap='viridis')

plt.title('Average Values of Key Features by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Average Value')
plt.legend(title='Features')
plt.xticks(rotation=0)
plt.show()
```
