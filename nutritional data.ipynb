```python
# Distribution of Calories
plt.figure(figsize=(10, 6))
sns.histplot(data['Calories'], bins=20, kde=True, color='blue')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count plot for Category
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Category',order=data['Category'].value_counts().index)
plt.title('Frequency of Items by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart for Category distribution
data['Category'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=sns.color_palette('pastel'))
plt.title('Category Proportion')
plt.ylabel('')
plt.show()
```

```python
correlation = data[['Calories', 'Total Fat', 'Protein']].corr()
print(correlation)

# Heatmap of correlations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Bar plot: Average Calories by Category
plt.figure(figsize=(12, 6))
data.groupby('Category')['Calories'].mean().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Average Calories by Category')
plt.ylabel('Average Calories')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#3.Time Series Plot: Calories over time
data['Date'] =pd.date_range(start='2023-01-01', periods=len( data), freq='D')  # Adding a Date column
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Date', y='Calories', color='green')
plt.title('Calories Over Time')
plt.xlabel('Date')
plt.ylabel('Calories')
plt.show()
```

```python
# Group by 'Category' and calculate mean for nutritional columns
category_nutrition = data.groupby('Category')[[
    'Saturated Fat', 'Saturated Fat (% Daily Value)', 'Trans Fat',
    'Cholesterol', 'Cholesterol (% Daily Value)', 'Sodium', 
    'Sodium (% Daily Value)', 'Carbohydrates', 'Carbohydrates (% Daily Value)',
    'Dietary Fiber', 'Dietary Fiber (% Daily Value)', 'Sugars', 
    'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 
    'Calcium (% Daily Value)', 'Iron (% Daily Value)'
]].mean()

# Display the result
category_nutrition
```

```python
# Plotting average values of selected nutrition metrics by Category
plt.figure(figsize=(15, 8))
category_nutrition[['Saturated Fat', 'Trans Fat', 'Cholesterol', 'Sodium', 'Carbohydrates']].plot(kind='bar', colormap='viridis', edgecolor='black')
plt.title('Average Nutritional Data by Category')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# List of nutritional features to plot
nutritional_features = [
    'Calories', 'Calories from Fat', 'Total Fat', 'Total Fat (% Daily Value)', 'Saturated Fat', 
    'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol', 'Cholesterol (% Daily Value)', 
    'Sodium', 'Sodium (% Daily Value)', 'Carbohydrates', 'Carbohydrates (% Daily Value)', 
    'Dietary Fiber', 'Dietary Fiber (% Daily Value)', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 
    'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)'
]

# Set up the figure for line plots
plt.figure(figsize=(15, 10))

# Iterate through each nutritional feature and create a line plot
for feature in nutritional_features:
    if feature in category_nutrition.columns:
        sns.lineplot(data=category_nutrition, x='Category', y=feature, label=feature, marker='o')

# Title and labels
plt.title('Line Plot of Nutritional Features by Category')
plt.xlabel('Category')
plt.ylabel('Nutritional Value')
plt.xticks(rotation=90)  # Rotate category labels to avoid overlap
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1))  # Adjust legend position

# Display the plot
plt.tight_layout()  # Adjust the layout to avoid clipping
plt.show()
```
