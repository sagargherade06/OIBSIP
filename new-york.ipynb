```python
import numpy as np

# Function to detect and handle outliers
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    print(f"Column: {column}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    
    # Filter out the outliers
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    print(f"Number of outliers removed: {len(df) - len(filtered_df)}")
    return filtered_df

# Apply to relevant columns
columns_to_check = ['price', 'minimum_nights', 'reviews_per_month', 'availability_365']
for column in columns_to_check:
    df = handle_outliers(df, column)
```

```python
def handle_outliers(df, column, multiplier=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    
    print(f"Column: {column}")
    print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    print(f"Number of outliers removed: {len(df) - len(filtered_df)}")
    return filtered_df

# Apply with adjusted multiplier
for column in columns_to_check:
    df = handle_outliers(df, column, multiplier=3)
```

```python
def handle_outliers_with_domain_knowledge(df, column, lower=None, upper=None):
    if lower is not None:
        df = df[df[column] >= lower]
    if upper is not None:
        df = df[df[column] <= upper]
    return df

# Example thresholds
df = handle_outliers_with_domain_knowledge(df, 'price', lower=1, upper=5000)
df = handle_outliers_with_domain_knowledge(df, 'minimum_nights', lower=1, upper=30)
```
